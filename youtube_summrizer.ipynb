```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Actice cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=['Confirmed', 'Deaths', 'Cured'], index='State/UnionTerritory', aggfunc = max)
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = 'Confirmed', ascending=False)
```

```python
statewise.style.background_gradient(cmap='cividis')
```

```python
top_15_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Date']]
top_15_active_cases = top_15_active_cases.sort_values(by='Active_Cases', ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(14,8))
plt.title('top 15 states with most active cases in India', size=20)
```

```python
ax = sns.barplot(data = top_15_active_cases.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'orange')
```

```python
top_15_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases', 'Deaths']]
top_15_active_cases = top_15_active_cases.sort_values(by='Active_Cases', ascending=False).head(15).reset_index()

plt.figure(figsize=(14, 8))
plt.title('Top 15 States with Most Active COVID-19 Cases in India', size=20)

ax = sns.barplot(data=top_15_active_cases,
                 y='State/UnionTerritory',
                 x='Active_Cases',
                 palette='hsv',
                 linewidth=2,
                 edgecolor='black')

for index, value in enumerate(top_15_active_cases['Active_Cases']):
    plt.text(value + 500, index, str(value), color='black', ha='left', fontsize=10)

plt.xlabel('Total Active Cases', fontsize=14)
plt.ylabel('States/Union Territories', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
top_15_death_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']]
top_15_death_cases = top_15_death_cases.sort_values(by='Deaths', ascending=False).head(15).reset_index()

plt.figure(figsize=(14, 8))
plt.title('Top 15 States with Most Death Cases in India', size=20)

ax = sns.barplot(data=top_15_death_cases,
                 y='State/UnionTerritory',
                 x='Deaths',
                 palette='hsv',
                 linewidth=2,
                 edgecolor='black')

for index, value in enumerate(top_15_death_cases['Deaths']):
    plt.text(value + 500, index, str(value), color='black', ha='left', fontsize=10)

plt.xlabel('Total Death Cases', fontsize=14)
plt.ylabel('States/Union Territories', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
testing_df.isnull().sum()
```

```python
testing_df.dropna(subset=['Negative', 'Positive'], inplace=True)
```

```python
testing_df['Negative'].fillna(method='ffill', inplace=True)
testing_df['Positive'].fillna(method='bfill', inplace=True)
```

```python
state_mapping = {
    'Andaman and Nicobar Islands': 'Andaman & Nicobar',
    'NCT of Delhi': 'Delhi'
}

covid_df['State/UnionTerritory'] = covid_df['State/UnionTerritory'].replace(state_mapping)
vaccine_df['State'] = vaccine_df['State'].replace(state_mapping)
testing_df['State'] = testing_df['State'].replace(state_mapping)
```

```python
vaccine_df.head()
```

```python
vaccine_df['Vaccine_Date'] = pd.to_datetime(vaccine_df['Vaccine_Date'])
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'])
```

```python
testing_df['Date'] = pd.to_datetime(testing_df['Date'])
```

```python
merged_df = pd.merge(covid_df, vaccine_df,
                     how='inner',
                     left_on=['Date', 'State/UnionTerritory'],
                     right_on=['Vaccine_Date', 'State'])
```

```python
final_df = pd.merge(merged_df, testing_df,
                    how='inner',
                    left_on=['Date', 'State/UnionTerritory'],
                    right_on=['Date', 'State'])
```

```python
final_df.drop(columns=['State_y'], inplace=True, errors='ignore')
```

```python
print(final_df.head())
```

```python
print(final_df.isnull().sum())
```

```python
columns_to_drop = [
    'Sputnik V (Doses Administered)', 'AEFI',
    '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)',
    '18-44 Years(Individuals Vaccinated)', '45-60 Years(Individuals Vaccinated)', '60+ Years(Individuals Vaccinated)',
    'Male(Individuals Vaccinated)', 'Female(Individuals Vaccinated)', 'Transgender(Individuals Vaccinated)',
    'Total Individuals Vaccinated'
]

final_df.drop(columns=columns_to_drop, inplace=True)
```

```python
correlation_matrix_pearson = final_df[['Active_Cases', 'Total Doses Administered', 'TotalSamples', 'Positive', 'Negative']].corr(method='pearson')
correlation_matrix_spearman = final_df[['Active_Cases', 'Total Doses Administered', 'TotalSamples', 'Positive', 'Negative']].corr(method='spearman')
print("Pearson Correlation Coefficient:\n", correlation_matrix_pearson)
print("\nSpearman Correlation Coefficient:\n", correlation_matrix_spearman)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(correlation_matrix_pearson, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Pearson Correlation (Linear Relationships)")
plt.subplot(1, 2, 2)
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Spearman Correlation (Non-Linear Relationships)")
plt.show()
```

```python
correlation_matrix_spearman = final_df[['Active_Cases', 'Total Doses Administered', 'TotalSamples', 'Positive', 'Negative']].corr(method='spearman')
print("Spearman Correlation Coefficient:\n", correlation_matrix_spearman)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Spearman Correlation (Non-Linear Relationships)")
plt.show()
```
